```python
# 이진 검색 하기
from typing import List

def search(nums: List[int], target: int) -> int:
    if not nums:
        return -1

    n = len(nums)
    left = 0
    right = n

    while left <= right:
        mid = (left + right) // 2 # int로 나눠떨어지지 않을 수 있음, // 쓰면 int로 나눠떨어짐

        if nums[mid] == target:
            return mid

        if nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid - 1
    return -1

arr = [2, 4, 5, 9, 11, 24, 70, 101]
print(-1 !=search(arr, 5), ",Index location : ", search(arr,5))
```

```python
# 회전된 배열에서 값 찾기, 모르면 print문 찍어서 값 확인하기

from typing import List

def search(nums: List[int], target: int) -> int:
    l = 0
    r = len(nums) - 1

    while l <= r:
        m = (l + r) // 2

        if nums[m] == target:
            return m

        if nums[l] <= nums[m]:
            if nums[l] <= target < nums[m]:
                r = m - 1
            else:
                l = m + 1
        else:
            if nums[m] < target <= nums[r]:
                l = m + 1
            else:
                r = m - 1
    return -1

nums = [7, 8, 1, 2, 3, 4, 5, 6]
target = 1
print(search(nums, target))
```

```python
# 이진 트리 활용한 문제 풀기
# tr.val = 루트노드
# tr.right.val = 왼쪽노드 ...?
# tr.left.val = 오른쪽노드
# 이진트리 세팅 코드
# pritn(seq)

import ast
import collections
class TreeNode: # 현재값을 반환하거나, 오른쪽 왼쪽 반환
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None

def deserialize(seq):
    if not seq :
        return None
    # print(seq.replace('null', 'None')) - 출력 확인용 코드
    seq = ast.literal_eval(seq.replace('null', 'None'))
    # ast.literal_eval - 파이썬에서 null값은 의미가 없기 때문에 스트링값을 null에서 None으로 바꿔준다
    root =  TreeNode(seq[0])
    q = collections.deque([root])
    idx = 1
    while q and idx < len(seq):
        node = q.popleft()
        node.left = TreeNode(seq[idx]) if seq[idx] else None
        idx += 1
        if node.left:
            q.append(node.left)
        if idx < len(seq) :
            node.right = TreeNode(seq[idx]) if seq[idx] else None
            idx += 1

            if node.right:
                q.append(node.right)
    return root

    # 5가 루트, deque에 5를 넣음
    # -> deque[5] (deque[root])
    # 11이 인덱스 1번이 됨
    # q.popleft() - 5를 가져옴
    # node(5)의 left를 11로 넣어줘라
    # 인덱스값+1 == 2가 됨, 15
    # q(deque)에 11넣기
    # node(11)의 right를 15로 넣어줘라
    # 인덱스값+1 == 1이됨
    # q.popleft() - 11을 가져옴
    # ...
    # 트리 만들기가 끝난 후 deque에는 1, 2, 6이 남아있음
tr = deserialize('[5, 11, 15, 1, 2, null, 6]')
print(tr.val)
print(tr.right.val)
print(tr.left.val)
print(tr.left.left.val)
```

```python
# 이진트리 맛보기 : 단계 순위 순회로 역방문 하기

import collections
from typing import List
import os
import sys

def do_level_order_from_bottom(root: List[TreeNode]) -> List[List[int]]:
    if not root:
        return []

    q = collections.deque([root])
    lines = []

    while q:
        nq = collections.deque()
        line = []
        while q:
            node = q.popleft()
            line += node.val, # +=을 사용하고 뒤에 ,를 넣으면서 반복되는 구조를 만들수 있음 ex) [11,15] (+=)
            if node.left:
                nq += node.left,
            if node.right:
                nq += node.right,
            q=nq
    return lines

res = do_level_order_from_bottom(deserialize('[5, 11, 15, 1, 2, null, 6]'))
print(res)
```
