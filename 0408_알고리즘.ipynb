```python
# 이진 검색 하기
from typing import List

def search(nums: List[int], target: int) -> int:
    if not nums:
        return -1

    n = len(nums)
    left = 0
    right = n

    while left <= right:
        mid = (left + right) // 2 # int로 나눠떨어지지 않을 수 있음, // 쓰면 int로 나눠떨어짐

        if nums[mid] == target:
            return mid

        if nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid - 1
    return -1

arr = [2, 4, 5, 9, 11, 24, 70, 101]
print(-1 !=search(arr, 5), ",Index location : ", search(arr,5))
```

```python
# 회전된 배열에서 값 찾기, 모르면 print문 찍어서 값 확인하기

from typing import List

def search(nums: List[int], target: int) -> int:
    l = 0
    r = len(nums) - 1

    while l <= r:
        m = (l + r) // 2

        if nums[m] == target:
            return m

        if nums[l] <= nums[m]:
            if nums[l] <= target < nums[m]:
                r = m - 1
            else:
                l = m + 1
        else:
            if nums[m] < target <= nums[r]:
                l = m + 1
            else:
                r = m - 1
    return -1

nums = [7, 8, 1, 2, 3, 4, 5, 6]
target = 1
print(search(nums, target))
```
