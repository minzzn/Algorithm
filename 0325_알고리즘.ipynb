```python
## 홀수와 짝수로 정렬하기

from typing import List
# 파라미터의 형식을 명시적으로 표기하기 위한
# 라이브러리

nums = [5, 3, 8, 12]

# ":" : 입력 파라미터가 리스트일 것이다
# "->" 반환 값이 리스트 일 것이다. 주석역할
# List[int] 형태로 반환할 것
def sort_even_odd(nums: List[int]) -> List[int]:
    l = 0
    r = 0

    while r < len(nums):
        if 0 == nums[r] % 2:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
        r += 1

    return nums

print([8, 12, 5, 3] == sort_even_odd(nums))
```

```python
## 홀수와 짝수로 정렬하기
## 짝수, 홀수 기준으로 오름차순 sorting을 한다면?

from typing import List
# 파라미터의 형식을 명시적으로 표기하기 위한
# 라이브러리

nums = [5, 3, 8, 12]

# ":" : 입력 파라미터가 리스트일 것이다
# "->" 반환 값이 리스트 일 것이다. 주석역할
# List[int] 형태로 반환할 것
def sort_even_odd(nums: List[int]) -> List[int]:
    l = 0
    r = 0

    while r < len(nums):
        if 0 == nums[r] % 2:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
        r += 1

    return nums, l

print(sort_even_odd(nums))

newList, left = sort_even_odd(nums)
sorted_sublist1 =  sorted(newList[:left])
sorted_sublist2 = sorted(newList[left:])
print(sorted_sublist1+sorted_sublist2)
```
