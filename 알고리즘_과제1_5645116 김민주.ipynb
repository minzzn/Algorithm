```python
# 1. 선택 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def selection_sort(input_list):
    for i in range(len(input_list)):
        min_index = i
        for j in range(i + 1, len(input_list)):
            if input_list[j][0] < input_list[min_index][0]:
                min_index = j
        input_list[i], input_list[min_index] = input_list[min_index], input_list[i]

selection_sort(times)

for time, inc in times:
    cnt += inc
    print(time,"-",cnt)
    mx = max(mx,cnt)

print("필요한 회의실 개수: ",mx)
```

```python
# 2. 버블

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def bubble_sort(input_list):
    for i in range(len(input_list) - 1, 0, -1):
        for j in range(i):
            if input_list[j][0] > input_list[j + 1][0]:
                input_list[j], input_list[j + 1] = input_list[j + 1], input_list[j]

bubble_sort(times)

for time, inc in times:
    cnt += inc
    print(time,"-",cnt)
    mx = max(mx,cnt)

print("필요한 회의실 개수: ",mx)
```

```python
# 3. 삽입 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def insertion_sort(input_list):
    for end in range(1, len(input_list)):
        for i in range(end, 0, -1):
            if input_list[i - 1][0] > input_list[i][0]:  # 시간을 기준으로 정렬
                input_list[i - 1], input_list[i] = input_list[i], input_list[i - 1]

insertion_sort(times)

# 회의실 개수 계산
for time, inc in times:
    cnt += inc
    print(time,"-",cnt)
    mx = max(mx, cnt)

print("필요한 회의실 개수:", mx)
```

```python
# 4. 병합 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def merge(arr, low, high):
    temp = []
    mid = (low + high) // 2
    i, j = low, mid + 1

    while i <= mid and j <= high:
        if arr[i][0] <= arr[j][0]:
            temp.append(arr[i])
            i += 1
        else:
            temp.append(arr[j])
            j += 1

    while i <= mid:
        temp.append(arr[i])
        i += 1

    while j <= high:
        temp.append(arr[j])
        j += 1

    for k in range(low, high + 1):
        arr[k] = temp[k - low]

    return arr


def merge_sort(arr, low, high):
    if (low < high):
        mid = (low + high) // 2

        merge_sort(arr, low, mid)
        merge_sort(arr, mid+1, high)

        merge(arr, low, high)

merge_sort(times, 0, len(times) - 1)

for time, inc in times:
    cnt += inc
    print(time,"-",cnt)
    mx = max(mx,cnt)

print("필요한 회의실 개수: ",mx)
```

```python
# 5. 퀵 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def quick_sort(array):
    # 리스트가 하나 이하의 원소만을 담고 있다면 종료
    if len(array) <= 1:
        return array

    pivot = array[0]    # 피벗은 첫 번째 원소
    tail = array[1:]    # 피벗을 제외한 리스트

    left_side = [x for x in tail if x[0] < pivot[0]]     # 분할된 왼쪽 부분
    right_side = [x for x in tail if x[0] >= pivot[0]]     # 분할된 오른쪽 부분

    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬을 수행하고, 전체 리스트를 반환
    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

# 시간을 기준으로 정렬
sorted_times = quick_sort(times)

# 회의실 개수 계산
for time, inc in sorted_times:
    cnt += inc
    print(time,"-",cnt)
    mx = max(mx, cnt)

print("필요한 회의실 개수:", mx)
```

```python
# 6. 힙 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def heapify(unsorted, index, heap_size):
    largest = index
    left = 2 * index + 1
    right = 2 * index + 2

    if left < heap_size and unsorted[left][0] > unsorted[largest][0]:
        largest = left

    if right < heap_size and unsorted[right][0] > unsorted[largest][0]:
        largest = right

    if largest != index:
        unsorted[largest], unsorted[index] = unsorted[index], unsorted[largest]
        heapify(unsorted, largest, heap_size)

def heap_sort(unsorted):
    n = len(unsorted)

    # 최대 힙을 구성
    for i in range(n // 2 - 1, -1, -1):
        heapify(unsorted, i, n)

    # 힙 정렬 수행
    for i in range(n - 1, 0, -1):
        unsorted[0], unsorted[i] = unsorted[i], unsorted[0]
        heapify(unsorted, 0, i)

    return unsorted

# 회의실 예약 시간대
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]

# 시작 시간과 종료 시간을 1과 -1로 변환하여 리스트 구성
times = []
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))



# 힙 정렬을 적용할 때 튜플의 첫 번째 값을 기준으로 정렬해야 합니다.
times_sorted = heap_sort(times)

# 수정된 부분: 회의실 개수 계산 시 오름차순으로 계산합니다.
# 회의실 개수 계산
for time, inc in times_sorted:
    cnt += inc
    mx = max(mx, cnt)
    print(time,"-",cnt)

print("필요한 회의실 개수:", mx)
# 다른 코드와 출력 결과가 다른 이유 - 힙 정렬의 특성상 최대 루트노드를 기준으로
# 정렬이 되고, 내림차순으로 리스트가 정렬이 되어 계산 과정에 문제가 생기는 것 같습니다.
# 계산을 해본 결과, 정렬한 리스트를 오름차순으로 바꾸어 계산하면 올바른 결과가 나왔지만,
# 코드로는 잘 구현이 되지 않았던것 같습니다.
```

```python
# 7. 계수 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
cnt = 0
mx = 0
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def counting_sort(arr):
    # 가장 큰 시간을 찾기 위해 반복
    max_time = max(arr, key=lambda x: x[0])[0]
    count = [0] * (max_time + 1)

    # 시간대별로 빈도수를 세기
    for time, _ in arr:
        count[time] += 1

    # 누적 빈도수 계산
    for i in range(1, len(count)):
        count[i] += count[i - 1]

    # 정렬된 결과를 담을 배열 초기화
    sorted_arr = [0] * len(arr)

    # 원래 배열을 뒤에서부터 순회하며 정렬된 배열에 채워넣기
    for i in range(len(arr) - 1, -1, -1):
        time, inc = arr[i]
        sorted_arr[count[time] - 1] = arr[i]
        count[time] -= 1

    return sorted_arr

# 시간대를 계수 정렬을 사용하여 정렬
sorted_intervals = counting_sort(times)

# 결과 출력
for time, inc in sorted_intervals:
    cnt += inc
    print(time, "-", cnt)
    mx = max(mx, cnt)

print("필요한 회의실 개수:", mx)
```

```python
# 8. 버킷 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]
times = []

# 회의 시작 시간과 종료 시간을 추출하여 times 리스트에 저장
for start, end in intervals:
    times.append((start, 1))
    times.append((end, -1))

def bucket_sort(num):
    # 최대 시간과 최소 시간을 찾는다
    max_time = max(num, key=lambda x: x[0])[0]
    min_time = min(num, key=lambda x: x[0])[0]

    # 버킷을 담을 비어 있는 list를 하나 생성한다
    arr = [[] for _ in range(max_time - min_time + 1)]

    # 버킷에 각 요소를 넣는다
    for n in num:
        index = n[0] - min_time
        arr[index].append(n)

    # 시간대별로 회의실 예약 상황을 계산한다
    room_count = 0
    room_counts = []
    for bucket in arr:
        for time, inc in bucket:
            room_count += inc
            room_counts.append((time, room_count))

    return room_counts, min_time

room_counts, min_time = bucket_sort(times)

# 시간대별 사용중인 회의실 개수 출력
for time, count in room_counts:
    print(time, "-", count)

# 최대 사용중인 회의실 개수 출력
max_rooms = max(count for _, count in room_counts)
print("필요한 회의실 개수:", max_rooms)
```

```python
# 9. 기수 정렬

# 배열, 정렬
# 필요한 회의실 구하기

# 1번 문제의 예시 대신, 더 많은 회의 배열을 임의로 만들어서 사용
intervals = [(12, 16), (14, 18), (10, 15), (5, 10), (9, 12)]



def radixSort(intervals):
    buckets = [[] for _ in range(24)]  # 시간대별로 버킷을 생성합니다. 시간은 0부터 23시까지 있으므로 24개의 버킷이 필요합니다.

    for start, end in intervals:
        buckets[start].append((start, 1))  # 회의 시작 시간에는 회의실을 예약하는 시점으로 카운트를 증가
        buckets[end].append((end, -1))     # 회의 종료 시간에는 회의실을 해제하는 시점으로 카운트를 감소

    sorted_intervals = []
    cnt = 0
    mx = 0

    for bucket in buckets:
        for time, inc in bucket:
            cnt += inc
            mx = max(mx, cnt)
            sorted_intervals.append((time, cnt))

    return sorted_intervals, mx

sorted_intervals, max_rooms = radixSort(intervals)

# 정렬된 회의 시간 출력
for time, room_count in sorted_intervals:
    print(time, "-", room_count)

print("필요한 회의실 개수:", max_rooms)
```
